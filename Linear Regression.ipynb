```python
#use both sklearn and statsmodels to get the regression results
from sklearn import linear_model
import statsmodels.api as sm
```

```python
# From statsmodels first

import statsmodels.api as sm

X = df[['Interest_Rate','Unemployment_Rate']] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = df['Stock_Index_Price']

X=sm.add_constant(X) #You need to add the coefficient in, its not given
model=sm.OLS(Y,X).fit()
print_model=model.summary()
print(print_model)

predictions=model.predict(X)
print(X,predictions)# check the syntax documentation for this to get the data
```

```python
#Using sklearn
from sklearn import linear_model
X = df[['Interest_Rate','Unemployment_Rate']] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = df['Stock_Index_Price']

regr=linear_model.LinearRegression()
regr.fit(X,Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# prediction with sklearn
New_Interest_Rate = 2.75
New_Unemployment_Rate = 5.3
prediction= regr.predict([[New_Interest_Rate ,New_Unemployment_Rate]])
print(prediction)
```

```python
### Disney Dataset Analysis
from pathlib import Path
my_csv = Path("C:/Users/antra.labroo/Desktop/Life/Interview Prep/disney_movies_total_gross.csv")
gross = pd.read_csv(my_csv.resolve(), sep=',')
gross.head()

genre_dummies =  pd.get_dummies(data=gross["genre"],drop_first=True)

# Inspect genre_dummies
genre_dummies.head()

X = genre_dummies # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = gross['inflation_adjusted_gross']

X=sm.add_constant(X) #You need to add the coefficient in, its not given
model=sm.OLS(Y,X).fit()
print_model=model.summary()
print(print_model)

print(model.predict(X))
#Linear Regression Using StatsModel
```

```python
#https://www.kaggle.com/code/hamelg/python-for-data-27-linear-regression/notebook
regression_model = linear_model.LinearRegression()

# Train the model using the mtcars data
regression_model.fit(X = genre_dummies, 
                     y = gross['inflation_adjusted_gross'])

# Check trained model y-intercept
print(regression_model.intercept_)

# Check trained model coefficients
print(regression_model.coef_)
print(regression_model.predict(genre_dummies))
```
